```python
import pandas as pd
```

```python
pd.options.mode.chained_assignment = None
```

```python
rawSongs = pd.read_csv('data/ml-03-data-processing-songs-dataset.csv')
```

```python
indexesToDrop = []

for index, row in rawSongs.iterrows():
    if row.isna().sum()/rawSongs.shape[1] >= 0.5:
        indexesToDrop.append(index)

rawSongs.drop(rawSongs.index[indexesToDrop], inplace=True )
```

```python
for column in rawSongs:
    # print(f'{column} - {round(rawSongs[column].isna().sum()/rawSongs.shape[0], 3)}')
    if rawSongs[column].isna().sum()/rawSongs.shape[0] >= 0.5:
        rawSongs = rawSongs.drop(column, axis = 1)
```

```python
rawSongs['Length (Duration)'] = rawSongs['Length (Duration)'].str.replace(',', '').astype(float)
```

```python
index = 0

for column in rawSongs:
    if index >= 6 or index == 4:
        outliers = rawSongs[rawSongs[column] > rawSongs[column].mean() + 3 * rawSongs[column].std()]
        if (len(outliers) != 0):
            print(column)
            display(outliers)
    index += 1
    
```

```python
import calendar
```

```python
for index, row in rawSongs.iterrows():
    if not row['Month'].isnumeric():
        rawSongs.at[index,'Month'] = list(calendar.month_abbr).index(row['Month'])
```

```python
rawSongs['Date'] = pd.to_datetime(
    dict(
        year = rawSongs['Year'], 
        month = rawSongs['Month'], 
        day = 1
    )
)
```

```python
now = pd.to_datetime('now')

rawSongs['Age'] = (now - rawSongs['Date']).astype('<m8[Y]') 
```

```python
rawSongs = rawSongs.drop(columns = ['Year', 'Month', 'Date'])
```
